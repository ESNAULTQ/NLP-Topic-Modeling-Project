```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import KFold
import gensim
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora.dictionary import Dictionary

# Charger les données
dftrain = pd.read_csv("Train.csv")
X_train = dftrain['ABSTRACT']

# Initialiser le CountVectorizer
vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2, max_features=1000)
X_train_vec = vectorizer.fit_transform(X_train)

# Préparer les données pour gensim
texts = [doc.split() for doc in X_train]
dictionary = Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# Paramètres de validation croisée
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Liste pour stocker les scores de cohérence pour chaque pli
coherence_scores = []

# Validation croisée
for train_index, test_index in kf.split(X_train_vec):
    # Diviser les données en ensembles d'entraînement et de test
    X_train_fold, X_test_fold = X_train_vec[train_index], X_train_vec[test_index]
    
    # Entraîner le modèle LDA sur l'ensemble d'entraînement
    lda_model = LatentDirichletAllocation(n_components=5, random_state=42)
    lda_model.fit(X_train_fold)
    
    # Convertir les ensembles d'entraînement en formats nécessaires pour gensim
    texts_fold = [texts[i] for i in train_index]
    dictionary_fold = Dictionary(texts_fold)
    corpus_fold = [dictionary_fold.doc2bow(text) for text in texts_fold]
    
    # Créer un modèle LDA avec gensim pour calculer la cohérence
    lda_model_gensim = gensim.models.ldamodel.LdaModel(corpus=corpus_fold, num_topics=10, id2word=dictionary_fold, passes=10)
    
    # Calculer le score de cohérence pour ce pli
    coherence_model_lda = CoherenceModel(model=lda_model_gensim, texts=texts_fold, dictionary=dictionary_fold, coherence='c_v')
    coherence_lda = coherence_model_lda.get_coherence()
    
    # Ajouter le score de cohérence à la liste
    coherence_scores.append(coherence_lda)
    print(f"Coherence Score for fold: {coherence_lda}")

# Moyenne des scores de cohérence à travers les plis
mean_coherence = sum(coherence_scores) / len(coherence_scores)
print(f"Mean Coherence Score: {mean_coherence}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import train_test_split, GridSearchCV
import gensim
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora.dictionary import Dictionary

# Charger les données
dftrain = pd.read_csv("Train.csv")
X_train = dftrain['ABSTRACT']

# 1. Optimisation des paramètres du CountVectorizer
vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2, max_features=1000)
X_train_vec = vectorizer.fit_transform(X_train)

# Préparer les données pour gensim
texts = [doc.split() for doc in X_train]
dictionary = Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# Séparation des données en ensembles d'entraînement et de validation
X_train_vec_train, X_train_vec_val = train_test_split(X_train_vec, test_size=0.2, random_state=42)

# 2. Optimisation des hyperparamètres avec GridSearchCV (réduit)
search_params = {
    'n_components': [5, 10], 
    'learning_decay': [0.7, 0.9],
}

lda = LatentDirichletAllocation(max_iter=10, random_state=42)
model = GridSearchCV(lda, param_grid=search_params, cv=3)
model.fit(X_train_vec_train)

# Afficher les meilleurs paramètres trouvés
print(f"Best Model's Params: {model.best_params_}")
print(f"Best Log Likelihood Score: {model.best_score_}")

# Utiliser le modèle LDA avec les meilleurs paramètres
best_lda_model = model.best_estimator_

# Entraîner le modèle LDA sur l'ensemble des données d'entraînement
best_lda_model.fit(X_train_vec_train)

# Créer un modèle LDA avec gensim pour calculer la cohérence
lda_model_gensim = gensim.models.ldamodel.LdaModel(
    corpus=corpus, 
    num_topics=best_lda_model.n_components, 
    id2word=dictionary, 
    passes=10,  # Passes réduites pour optimiser le temps
    iterations=100  # Iterations réduites pour optimiser le temps
)

# Calculer le score de cohérence pour ce modèle
coherence_model_lda = CoherenceModel(model=lda_model_gensim, texts=texts, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()

print(f"Coherence Score: {coherence_lda}")
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import KFold, GridSearchCV
import gensim
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora.dictionary import Dictionary

# Charger les données
dftrain = pd.read_csv("Train.csv")
X_train = dftrain['ABSTRACT']

# 1. Optimisation des paramètres du CountVectorizer
vectorizer = CountVectorizer(stop_words='english', max_df=0.95, min_df=2, max_features=1000)
X_train_vec = vectorizer.fit_transform(X_train)

# Préparer les données pour gensim
texts = [doc.split() for doc in X_train]
dictionary = Dictionary(texts)
corpus = [dictionary.doc2bow(text) for text in texts]

# Paramètres de validation croisée
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Liste pour stocker les scores de cohérence pour chaque pli
coherence_scores = []

# 2. Optimisation des hyperparamètres avec GridSearchCV
search_params = {
    'n_components': [5, 10, 15], 
    'learning_decay': [0.5, 0.7, 0.9],
    'learning_offset': [10, 50, 100],
    'max_iter': [10, 20, 30]
}

# Modèle LDA
lda = LatentDirichletAllocation(random_state=42)

# Utiliser GridSearchCV pour trouver les meilleurs paramètres
model = GridSearchCV(lda, param_grid=search_params, cv=5)
model.fit(X_train_vec)

# Afficher les meilleurs paramètres trouvés
print(f"Best Model's Params: {model.best_params_}")
print(f"Best Log Likelihood Score: {model.best_score_}")

# Utiliser le modèle LDA avec les meilleurs paramètres
best_lda_model = model.best_estimator_

# Validation croisée avec le meilleur modèle
for train_index, test_index in kf.split(X_train_vec):
    # Diviser les données en ensembles d'entraînement et de test
    X_train_fold, X_test_fold = X_train_vec[train_index], X_train_vec[test_index]
    
    # Entraîner le modèle LDA sur l'ensemble d'entraînement
    lda_model = LatentDirichletAllocation(
        n_components=best_lda_model.n_components,
        learning_decay=best_lda_model.learning_decay,
        learning_offset=best_lda_model.learning_offset,
        max_iter=best_lda_model.max_iter,
        random_state=42
    )
    lda_model.fit(X_train_fold)
    
    # Convertir les ensembles d'entraînement en formats nécessaires pour gensim
    texts_fold = [texts[i] for i in train_index]
    dictionary_fold = Dictionary(texts_fold)
    corpus_fold = [dictionary_fold.doc2bow(text) for text in texts_fold]
    
    # Créer un modèle LDA avec gensim pour calculer la cohérence
    lda_model_gensim = gensim.models.ldamodel.LdaModel(
        corpus=corpus_fold, 
        num_topics=best_lda_model.n_components, 
        id2word=dictionary_fold, 
        passes=20,  # Augmenter le nombre de passes
        iterations=400  # Augmenter le nombre d'itérations
    )
    
    # Calculer le score de cohérence pour ce pli
    coherence_model_lda = CoherenceModel(model=lda_model_gensim, texts=texts_fold, dictionary=dictionary_fold, coherence='c_v')
    coherence_lda = coherence_model_lda.get_coherence()
    
    # Ajouter le score de cohérence à la liste
    coherence_scores.append(coherence_lda)
    print(f"Coherence Score for fold: {coherence_lda}")

# Moyenne des scores de cohérence à travers les plis
mean_coherence = sum(coherence_scores) / len(coherence_scores)
print(f"Mean Coherence Score: {mean_coherence}")
```
